# Question 1

```python
import matplotlib.pyplot as plt
import scipy
import cv2
import numpy as np
```

## Set up parameters

```python
size = 1000
sigma = 100
mid = int(size/2)
center_coord = (mid, mid)
thickness = -1
radius = 100
```

## Question 1A Generate all white background

```python
all_white = 255 * np.ones(shape=(size, size,), dtype=np.uint8)
white_bg_with_black_circ = cv2.circle(all_white, center_coord, radius, color=0, thickness=-1)
# Unclip it
unclipped = np.array(white_bg_with_black_circ, dtype=float)
```

```python
rslt = scipy.ndimage.gaussian_laplace(unclipped, sigma=sigma)

array = np.array(list(rslt))

print(sigma)
plt.plot(list(range(size)), array[mid])
plt.title("Center-slice of the image with sigma " + str(sigma)+ ' and radius ' + str(radius))
plt.show()

plt.imshow(rslt)
plt.title("Image with sigma " + str(sigma) + ' and radius ' + str(radius))
plt.show()
```

## Question 1B

```python
all_black = np.zeros(shape=(size, size,), dtype=np.uint8)

black_bg_with_white_circ = cv2.circle(all_black, center_coord, radius, color=255, thickness=-1)

unclipped = np.array(black_bg_with_white_circ, dtype=float)

rslt = scipy.ndimage.gaussian_laplace(unclipped, sigma=sigma)

array = np.array(list(rslt))

print(sigma)
plt.plot(list(range(size)), array[mid])
plt.title("Center-slice of the image with sigma " + str(sigma)+ ' and radius ' + str(radius))
plt.show()

plt.imshow(rslt)
plt.title("Image with sigma " + str(sigma) + ' and radius ' + str(radius))
plt.show()
```
